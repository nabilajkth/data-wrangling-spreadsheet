```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading data
df = pd.read_csv('/content/portugal_listinigs.csv')
```

```python
# Menampilkan 5 data teratas
df.head(5)
```

```python
# Cek Missing Value
df.isnull()
```

```python
# Informasi struktur data
df.info()
```

```python
# MENGUBAH TIPE DATA

# Mengubah tipe data kolom HasParking to bool
df['HasParking'] = df['HasParking'].astype(bool)

# Mengubah tipe data kolom Garage to bool
df['Garage'] = df['Garage'].astype(bool)

# Mengubah tipe data kolom Elevator to bool
df['Elevator'] = df['Elevator'].astype(bool)

# Mengubah tipe data kolom ElectricCarsCharging to bool
df['ElectricCarsCharging'] = df['ElectricCarsCharging'].astype(bool)

# Mengubah tipe data kolom PublishDate to datetime
df['PublishDate'] = pd.to_datetime(df['PublishDate'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Mengubah tipe data kolom ConstructionYear to datetime
df['ConstructionYear'] = pd.to_datetime(df['ConstructionYear'], format='%Y', errors='coerce')
print(df.dtypes)
```

```python
# Hitung total nilai yang hilang per kolom
missing_counts = df.isnull().sum()

# Hitung persentase nilai yang hilang per kolom
missing_percentage = (missing_counts / len(df)) * 100

# Gabungkan hasil menjadi DataFrame
missing_summary = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing Percentage (%)': missing_percentage
})

print(missing_summary)
```

```python
# Mengisi missing values dengan strategi yang ditentukan
df['Price'] = df['Price'].fillna(df['Price'].median())
df['Town'] = df['Town'].fillna(method='ffill')
df['Type'] = df['Type'].fillna(df['Type'].mode()[0])
df['EnergyCertificate'] = df['EnergyCertificate'].fillna('Unknown')
df['Parking'] = df['Parking'].fillna('None')
df['Elevator'] = df['Elevator'].fillna(False)
df['NumberOfBathrooms'] = df['NumberOfBathrooms'].fillna(df['NumberOfBathrooms'].median())

df['HasParking'] = df['HasParking'].fillna(False)
df['ConstructionYear'] = df['ConstructionYear'].fillna(df['ConstructionYear'].median())
df['TotalArea'] = df['TotalArea'].fillna(df['TotalArea'].median())
df['TotalRooms'] = df['TotalRooms'].fillna(df['TotalRooms'].median())
df['LivingArea'] = df['LivingArea'].fillna(df['LivingArea'].median())

# Ekstraksi angka dari kolom 'Floor' menggunakan str.extract
df['Floor'] = df['Floor'].str.extract('(\d+)').astype(float)

# Imputasi nilai missing dengan median
df['Floor'] = df['Floor'].fillna(df['Floor'].median())

df['GrossArea'] = df['GrossArea'].fillna(df['GrossArea'].median())
df['EnergyEfficiencyLevel'] = df['EnergyEfficiencyLevel'].fillna('Unknown')
df['Garage'] = df['Garage'].fillna('None')
df['ElectricCarsCharging'] = df['ElectricCarsCharging'].fillna(False)
df['NumberOfBedrooms'] = df['NumberOfBedrooms'].fillna(df['NumberOfBedrooms'].median())
df['NumberOfWC'] = df['NumberOfWC'].fillna(df['NumberOfWC'].median())
df['LotSize'] = df['LotSize'].fillna(df['LotSize'].median())


# Verifikasi hasil
df.isnull().sum()
```
